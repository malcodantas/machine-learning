<p>O Algoritmo K-means inicializa com a escolha randomica dos centroides; que sao base do funcionamento do algoritmo, porém isso é um problema, pois inicializações diferentes geram resultados diferentes, logo foi criando um algoritmo baseado no k-means que resolve esse problema, escolhendo de forma inteligente a inicializaçao dos centroids o <b>K-means++</b></p>
<ul>
    <i>conhecimento prévio do numero de clusters</i>
</ul>
<p>Se nao souber o numero de clusters pode usar a seguinte formular k=sqrt(N/2) onde N é o numero de registros ou ultiliza o <b>Elbow method</b> que tenta varios valores de K e nao existe garantia para encontrar o K correto</p>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.datasets.samples_generator import make_blobs # cria base de dados para teste
```

```python
x,y =make_blobs(n_samples=200,centers=4)
cores=['r.','g.','b.','y.']
plt.scatter(x[:,0],x[:,1])

```

```python
#Treinamento 
kmeans=KMeans(4)
kmeans.fit(x)
previsoes=kmeans.predict(x)
plt.scatter(x[:,0],x[:,1],c=previsoes)
```
